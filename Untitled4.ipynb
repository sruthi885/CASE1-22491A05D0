```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sumanth = pd.read_csv('energy.csv')
print(sumanth)

#LINE CHART ANDHRA KARNATAKA
srh=sumanth[sumanth['State']=='Andhra']
rcb=sumanth[sumanth['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='blue')
diag[1].plot(rcb['Date'],rcb['Consumption'],color='magenta')

srh=sumanth[sumanth['State']=='Telengana']
rcb=sumanth[sumanth['State']=='Tripura']
fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='lightgreen')
diag[1].bar(rcb['Date'],rcb['Consumption'],color='hotpink')

cse4='Kerala'
cse5='Delhi'
tdp=[cse4,cse5]
srh=sumanth[sumanth['State']=='Kerala']['Consumption'].sum()
rcb=sumanth[sumanth['State']=='Delhi']['Consumption'].sum()
ysr=[srh,rcb]
fig,diag=plt.subplots(2)
diag[0].pie(ysr,labels=tdp)
```
